## Chi-Square Results

Advanced Analysis: Statistical Tests
Perform statistical tests to determine if the differences in turnover rates are statistically significant.

Chi-Square Test for Categorical Variables
For categorical variables like BusinessTravel, Department, Gender, you can use the Chi-Square test:

```python
from scipy.stats import chi2_contingency


contingency_business_travel = pd.crosstab(df['BusinessTravel'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_business_travel)

print("\nChi-Square Test for Business Travel:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")

contingency_department = pd.crosstab(df['Department'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_department)

print("\nChi-Square Test for Department:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")

contingency_gender = pd.crosstab(df['Gender'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_gender)

print("\nChi-Square Test for Gender:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")

contingency_complaint_filed = pd.crosstab(df['complaintfiled'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_complaint_filed)

print("\nChi-Square Test for Complaint Filed:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")


contingency_salary_hike = pd.crosstab(df['PercentSalaryHike'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_salary_hike)

print("\nChi-Square Test for Salary Hike:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")


contingency_home_distance = pd.crosstab(df['DistanceFromHome'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_home_distance)

print("\nChi-Square Test for Distance From Home:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")


contingency_monthly_income = pd.crosstab(df['MonthlyIncome'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_monthly_income)

print("\nChi-Square Test for Monthly Income:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")


contingency_age = pd.crosstab(df['Age'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_age)

print("\nChi-Square Test for Age:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")


contingency_promo = pd.crosstab(df['YearsSinceLastPromotion'], df['Left'])
chi2, p, dof, ex = chi2_contingency(contingency_promo)

print("\nChi-Square Test for Years Since Last Promotion:")
print(f"Chi2 Stat: {chi2}, p-value: {p}")
```

### KNN Cross-Validation

```python
k_values = [i for i in range (1,31)]
scores = []

scaler = StandardScaler()
X = scaler.fit_transform(X)

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, X, y, cv=3)
    scores.append(np.mean(score))
```

```python
sns.lineplot(x = k_values, y = scores, marker = 'o')
plt.xlabel("K Values")
plt.ylabel("Accuracy Score")
```

We can see from our chart that k = 10+ all have an accuracy score of just under 84%. As these are pretty tied for the best score, it is advisable to use a smaller value for k. This is because when using higher values of k, the model will use more data points that are further away from the original... which seems to make it more accurate? Another option would be to explore other evaluation metrics.

```python
best_index = np.argmax(scores)
best_k = k_values[best_index]

knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
```

```python
from sklearn.svm import SVC
```

```python
def train_and_evalSVR(model, X, y):
    print('\nResults from algorithm {}:'.format(model))

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=44)
    
    svr = SVC(kernel='rbf', C=1.0)
    model = svr.fit(X_train, y_train)
    y_pred = svr.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    scores = cross_val_score(svr, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    rmse = np.sqrt(-np.mean(scores))
    sd = np.sqrt(np.var(scores))

    print("Mean Squared Error: {:.2f}".format(mse))
    print("R2 Score: {:.2f}".format(r2))
    print("Mean cross-validation RMSE is {:.3f} with SD {:.3f}".format(rmse, sd))

    
    return model
     
```

```python
svr_model = SVC(kernel='rbf', C=100, gamma=0.1)
_ = train_and_evalSVR(svr_model, X, y)
```

# <center> Results and Evaluation

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

pipeline = Pipeline([
    ('scaler', StandardScaler()), 
    ('logreg', LogisticRegression(penalty='elasticnet', solver='saga', max_iter=1000)) 
])

param_grid = {
    'logreg__C': [0.01, 0.1, 1, 10, 100],
    'logreg__l1_ratio': [0, 0.5, 1] 
}

grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train_poly, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_poly)
logreg_grid_score = accuracy_score(y_test, y_pred)

print("Best Parameters: ", grid_search.best_params_)
print("Accuracy: ", logreg_grid_score)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=42)

param_grid_rf = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='accuracy')
grid_search_rf.fit(X_train, y_train)

best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
rf_grid_score = accuracy_score(y_test, y_pred_rf)

print("Best Parameters for RF: ", grid_search_rf.best_params_)
print("RF Accuracy: ", rf_grid_score)
print(classification_report(y_test, y_pred_rf))
```
